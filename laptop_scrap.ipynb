```python
import requests
from bs4 import BeautifulSoup
```

```python
laptop_availeble = True
laptops_details = []
page_no = 1

while laptop_availeble:
    url = f"https://www.dell.com/en-us/shop/dell-laptops/sr/laptops?page={page_no}"  
    response = requests.get(url)
    # print(f"Response Status : {response.status_code}")
    page_no += 1
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        ps_wrapper = soup.find('div', class_='dell-ps ps', id='ps-wrapper')
        if not ps_wrapper:
            laptop_availeble = False
            print("No laptops available on this page. Exiting loop.")
            break  
        
        product_sections = soup.find_all('section', class_='ps-show-hide')
        if not product_sections:
            laptop_availeble = False
            print("No product sections found. Exiting loop.")
            break  
        
        for section in product_sections:
            title_tag = section.find('h3', class_='ps-title')
            laptop_name = title_tag.get_text(strip=True) if title_tag else "Unknown Laptop"
            
            processor_value = "Unknown Processor"
            specs_container = section.find('div', class_='ps-variant-specs')
            if specs_container:
                specs = specs_container.find_all('li', class_='variant-specs-column')
                for spec in specs:
                    labels = spec.find_all('div', class_='spec-label')
                    values = spec.find_all('div', class_='spec-value')
                    for label, value in zip(labels, values):
                        if 'Processor' in label.get_text(strip=True):
                            processor_value = value.get_text(strip=True)
                            break

            laptop_detail = {
                'laptop': laptop_name,
                'processor': processor_value
            }
            laptops_details.append(laptop_detail)
        
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        laptop_availeble = False  

print(f"Laptop details collected: {laptops_details}")
```
